###Card class

Below is a completed version of the `Card` class from HOA 8.1. The rest of the game depends on this class, and you should not modify it.

```
class Card:
  """
  A class representing a single card in a deck of playing cards.

  Attributes
  ----------
  __suit : int
    One of [0, 1, 2, 3], where 0 is clubs, 1 is diamonds, 2 is hearts, and 3
    is spades
  __rank : int
    One of [2, 3, 4, ..., 9, 10, 11, 12, 13, 14], where 11 means a Jack, 12 means
    a Queen, 13 a King, and 14 an Ace
  """

  def __init__(self, suit, rank):
    """
    Construct a playing card.

    parameters
    ----------
    suit : int
      One of [0, 1, 2, 3], where 0 is clubs, 1 is diamonds, 2 is hearts, and 3
      is spades
    rank : int
      One of [2, 3, 4, ..., 9, 10, 11, 12, 13, 14], where 11 means a Jack, 12 means
      a Queen, 13 a King, and 14 an Ace
    """
    self.__suit = suit
    self.__rank = rank

  def getSuit(self):
    """
    Return the suit of this card, as an integer in [0, 1, 2, 3]; 0 is clubs, 
    1 is diamonds, 2 is hearts, and 3 is spades
    """
    return self.__suit

  def setSuit(self, newSuit):
    """
    Change the suit of this card.

    Parameters
    ----------
    newSuit : int
      One of [0, 1, 2, 3], where 0 is clubs, 1 is diamonds, 2 is hearts, and 3
      is spades. If newSuit is none of these, no change is made to the Card.
    """
    if 0 <= newSuit <= 3:
      self.__suit = newSuit

  def getRank(self):
    """
    Return the rank of this card, as an integer in [2, ..., 14], where 2 
    through 10 are pip counts and 11 means a Jack, 12 means a Queen, 13 a King, 
    and 14 an Ace
    """
    return self.__rank

  def setRank(self, newRank):
    """
    Change the rank of this card.

    Parameters
    ----------
    newRank : int
      One of [2, 3, 4, ..., 9, 10, 11, 12, 13, 14], where 11 means a Jack, 
      12 means a Queen, 13 a King, and 14 an Ace. If newRank is none of these,
      no change is made to the Card.
    """
    if 2 <= newRank <= 14:
      self.__rank = newRank

  def __eq__(self, otherCard):
    """
    Determine if this Card is equivalent to another Card.

    Parameters
    ----------
    otherCard : Card
      Card to compare this one to

    Returns
    -------
      True if the suits and ranks are the same, false otherwise.
    """
    return self.__suit == otherCard.__suit and self.__rank == otherCard.__rank

  def __lt__(self, otherCard):
    """
    Determine if this Card is less than another Card.

    Parameters
    ----------
    otherCard : Card
      Card to compare this one to

    Returns
    -------
      If the ranks are not the same, return result of 
      self.__rank < otherCard.__rank. Otherwise, return result of 
      self.__suit < otherCard.__suit. 
    """
    if self.__rank != otherCard.__rank:
      return self.__rank < otherCard.__rank
    else:
      return self.__suit < otherCard.__suit

  def __str__(self):
    """
    Return a string representation of this Card, like 'Ace of Spades'. 
    """
    ranks = [0, 0, '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 
             'Queen', 'King', 'Ace']
    suits = ['Clubs', 'Diamonds', 'Hearts', 'Spades']

    return '{0:s} of {1:s}'.format(ranks[self.__rank], suits[self.__suit])
```
