```
class Planet:
  """
  A class representing a planet in a model of the solar system.

  Attributes
  ----------
  name : str 
    name of the planet
  radius : float
    radius of the planet in km
  mass : float
    mass of the planet in kg
  distance : float
    distance between planet and Sun, in km
  """

  def __init__(self, iName, iRad, iM, iDist):
    """
    Constructs a Planet object with the specified attributes.

    Parameters 
    ----------
    iName : str
      name of the planet
    iRad : float
      radius of the planet in km
    iM : float
      mass of the planet in kg
    iDist : float
      distance between planet and Sun, in km
    """
    self.__name = iName
    self.__radius = iRad
    self.__mass = iM 
    self.__distance = iDist

  def getName(self):
    """
    Get the planet's name.
    """
    return self.__name

  def setName(self, newName):
    """
    Change the name of the planet.

    Parameters
    ----------
    newName : str
      New name for the planet.
    """
    self.__name = newName

  def getRadius(self):
    """
    Get the planet's radius, in km.
    """
    return self.__radius

  def setRadius(self, newRadius):
    """
    Change the radius of the planet.

    Parameters
    ----------
    newRadius : float
      New radius for the planet, in km.
    """
    self.__radius = newRadius

  def getMass(self):
    """
    Get the planet's mass, in kg. 
    """
    return self.__mass

  def setMass(self, newMass):
    """
    Change the mass of the planet.

    Parameters
    ----------
    newMass : float
      New mass for the planet, in kg.
    """
    self.__mass = newMass

  def getDistance(self):
    """
    Get the planet's distance from the Sun, in km.
    """
    return self.__distance

  def setDistance(self, newDistance):
    """
    Change the planet's distance from the Sun.

    Parameters
    ----------
    newDistance : float
      New distance from the Sun, in km.
    """
    self.__distance = newDistance

  def getVolume(self):
    """
    Get the planet's volume, in km^3.
    """
    import math
    v = 4 / 3 * math.pi * self.__radius ** 3
    return v

  def getSurfaceArea(self):
    """
    Get the planet's surface area, in km^2.
    """
    import math
    sa = 4 * math.pi * self.__radius ** 2
    return sa 

  def getDensity(self):
    """
    Get the planet's density, in kg / km^3.
    """
    d = self.__mass / self.getVolume()
    return d
```

```
class Planet:
  """
  A class representing a planet in a model of the solar system.

  Attributes
  ----------
  name : str 
    name of the planet
  radius : float
    radius of the planet in km
  mass : float
    mass of the planet in kg
  distance : float
    distance between planet and Sun, in km
  """

  def __init__(self, iName, iRad, iM, iDist):
    """
    Constructs a Planet object with the specified attributes.

    Parameters 
    ----------
    iName : str
      name of the planet
    iRad : float
      radius of the planet in km
    iM : float
      mass of the planet in kg
    iDist : float
      distance between planet and Sun, in km
    """
    self.__name = iName
    self.__radius = iRad
    self.__mass = iM 
    self.__distance = iDist

  def getName(self):
    """
    Get the planet's name.
    """
    return self.__name

  def setName(self, newName):
    """
    Change the name of the planet.

    Parameters
    ----------
    newName : str
      New name for the planet.
    """
    self.__name = newName

  def getRadius(self):
    """
    Get the planet's radius, in km.
    """
    return self.__radius

  def setRadius(self, newRadius):
    """
    Change the radius of the planet.

    Parameters
    ----------
    newRadius : float
      New radius for the planet, in km.
    """
    self.__radius = newRadius

  def getMass(self):
    """
    Get the planet's mass, in kg. 
    """
    return self.__mass

  def setMass(self, newMass):
    """
    Change the mass of the planet.

    Parameters
    ----------
    newMass : float
      New mass for the planet, in kg.
    """
    self.__mass = newMass

  def getDistance(self):
    """
    Get the planet's distance from the Sun, in km.
    """
    return self.__distance

  def setDistance(self, newDistance):
    """
    Change the planet's distance from the Sun.

    Parameters
    ----------
    newDistance : float
      New distance from the Sun, in km.
    """
    self.__distance = newDistance

  def getVolume(self):
    """
    Get the planet's volume, in km^3.
    """
    import math
    v = 4 / 3 * math.pi * self.__radius ** 3
    return v

  def getSurfaceArea(self):
    """
    Get the planet's surface area, in km^2.
    """
    import math
    sa = 4 * math.pi * self.__radius ** 2
    return sa 

  def getDensity(self):
    """
    Get the planet's density, in kg / km^3.
    """
    d = self.__mass / self.getVolume()
    return d

  def __str__(self):
    """
    Get the name of this planet as a string.
    """
    return self.__name

  def __lt__(self, otherPlanet):
    """
    Compare this planet's distance to the Sun with that of another planet.

    Parameters
    ----------
    otherPlanet : Planet
      Planet to compare this planet's distance to.

    Returns
    -------
    True if this planet's distance to the Sun is less than the other planet's,
    False otherwise.
    """
    return self.__distance < otherPlanet.__distance

  def __gt__(self, otherPlanet):
    """
    Compare this planet's distance to the Sun with that of another planet.

    Parameters
    ----------
    otherPlanet : Planet
      Planet to compare this planet's distance to.

    Returns
    -------
    True if this planet's distance to the Sun is greater than the other planet's,
    False otherwise.
    """
    return self.__distance > otherPlanet.__distance
```

### (Extended) Listing 10.7

```
class Sun:
    """
    A class representing the Sun in a simulation of the solar system.

    Attributes
    ----------
    name : str
        Name of the Sun.
    radius : float
        Radius of the Sun, in km.
    mass : float
        Mass of the Sun, in kg.
    temp : float
        Temperature of the Sun, in degrees K.
    """

    def __init__(self, iName, iRad, iM, iTemp):
        """
        Constructs a Sun object with the specified attributes.

        Parameters
        ----------
        iName : str
            Name of the Sun.
        iRad : float
            Radius of the Sun, in km.
        iM : float
            Mass of the Sun, in kg.
        iTemp : float
            Temperature of the Sun, in degrees K.
        """
        self.__name = iName
        self.__radius = iRad
        self.__mass = iM
        self.__temp = iTemp

    def getMass(self):
        """
        Get the mass of the Sun in kg.
        """
        return self.__mass

    def __str__(self):
        """
        Get the name of the Sun as a string.
        """
        return self.__name
```

## (Extended) Listing 10.8

The cell below shows the full `SolarSystem` class, including class and method docstrings.

```
class SolarSystem:
    """
    Class representing the solar system in a simulation.

    Attributes
    ----------
    theSun : Sun
        The star at the center of the solar system.
    planets : list of Planet
        The planets in the solar system.
    """

    def __init__(self, aSun):
        """
        Construct a solar system object with the specified attribute.

        Parameters
        ----------
        aSun : Sun
            The star at the center of the solar system.
        """
        self.__theSun = aSun
        self.__planets = []

    def addPlanet(self, aPlanet):
        """
        Add a planet to the solar system.

        Parameters
        ----------
        aPlanet : Planet
            Planet to add to the solar system.
        """
        self.__planets.append(aPlanet)

    def showPlanets(self):
        """
        Print the names of the planets in the solar system.
        """
        for aPlanet in self.__planets:
            print(aPlanet)
```

##Session 10.11 Creating and showing a solar system with four planets

```
sun = Sun('Sun', 5000, 1000, 5800)
ss = SolarSystem(sun)
```

```
p = Planet('Mercury', 19, 10, 25)
ss.addPlanet(p)

# it's not necessary to have the name p for the planets
ss.addPlanet(Planet('Earth', 50, 60, 30))

ss.addPlanet(Planet('Mars', 47, 50, 35))

ss.addPlanet(Planet('Jupiter', 75, 100, 50))

ss.showPlanets()
```
