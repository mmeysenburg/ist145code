##Listing 3.1 Helper Functions Map Characters to Numbers

```
def letterToIndex(letter):
  """
  Convert a lowercase letter, or a space, into a number.

  parameters
  ----------
  letter : string
    Single, lowercase letter to convert
  
  returns
  -------
    Number corresponding to the letter: 'a' is 0, 'b' is 1, etc.
    A space is 26
  """
  from string import ascii_lowercase
  alphabet = ascii_lowercase + ' '
  idx = alphabet.find(letter)
  if idx == -1:
    print("error:", letter, "is not in the alphabet")
  return idx

def indexToLetter(idx):
  """
  Convert an integer into a lowercase letter or space.

  parameters
  ----------
  idx : int
    Integer in the closed range [0, 26]
  
  returns
  -------
    Letter between 'a' (0) and 'z' (25), or a space (26).
  """
  from string import ascii_lowercase
  alphabet = ascii_lowercase + ' '
  letter = ''
  if idx >= len(alphabet):
    print('error:', idx, 'is too large')
  elif idx < 0:
    print('error:', idx, 'is less than 0')
  else:
    letter = alphabet[idx]
  return letter
```

##Listing 3.2 Scrambling a plaintext message

```
def scramble2Encrypt(plainText):
  """
  Encrypt plaintext using a two-rail rail-fence cipher.

  parameters
  ----------
  plaintext : string
    The plaintext to encrypt.

  returns
  -------
    The ciphertext produced with the rail-fence cipher.
  """
  evenChars = ''
  oddChars = ''
  charCount = 0
  for ch in plainText:
    if charCount % 2 == 0:
      evenChars = evenChars + ch
    else:
      oddChars = oddChars + ch
    charCount = charCount + 1
  cipherText = oddChars + evenChars
  return cipherText
```

##Listing 3.3 Decrypting a transposed message

```
def scramble2Decrypt(cipherText):
  """
  Decrypt text ciphered with a two-rail rail-fence cipher.

  parameters
  ----------
  cipherText : string
    String of ciphertext created with a two-rail fail-fence cipher

  returns
  -------
    Decrypted plaintext
  """
  halfLength = len(cipherText) // 2
  evenChars = cipherText[halfLength:]
  oddChars = cipherText[:halfLength]
  plainText = ''

  for i in range(halfLength):
    plainText = plainText + evenChars[i]
    plainText = plainText + oddChars[i]

  if len(oddChars) < len(evenChars):
    plainText = plainText + evenChars[-1]

  return plainText
```

##Listing 3.5 Encrypting a message using a substitution cipher

```
def substitutionEncrypt(plainText, key):
  """
  Encrypt a string using the substitution cipher.

  parameters
  ----------
  plainText : string
    Plaintext to encrypt; assumed to be only letters and spaces
  key : string
    27-characacter key; scrambled lowercase letter and a space
  """
  from string import ascii_lowercase
  alphabet = ascii_lowercase + ' '
  plainText = plainText.lower()
  cipherText = ''
  for ch in plainText:
    idx = alphabet.find(ch) 
    cipherText = cipherText + key[idx]
  return cipherText
```

##Listing 3.6 Removing one character from a string

```
def removeChar(string, idx):
  """
  Remove a character from a string.

  parameters
  ----------
  string : string
    String to remove a character from
  idx : int
    Index of character to remove; assumed to be a valid index for the string

  returns
  -------
    A new version of the string with the specified character removed.
  """
  return string[0:idx] + string[idx + 1:]
```

##Listing 3.7 Function to generate a random key from the alphabet

```
def keyGen():
  """
  Generate a random key for use in the substitution cipher.

  returns
  -------
    A randomly-perumuted string containing all lowercase letters and a space.
  """
  import random
  from string import ascii_lowercase
  alphabet = ascii_lowercase + ' '
  key = ''
  for i in range(len(alphabet) - 1, -1, -1):
    idx = random.randint(0, i)
    key = key + alphabet[idx]
    alphabet = removeChar(alphabet, idx)
  return key
```

##Listing 3.8 Removing duplicate letters from a string

```
def removeDupes(myString):
  """
  Function to remove duplicate characters from a string.

  parameters
  ----------
  myString : string
    String to rid of duplicate characters

  returns
  -------
    A new version of myString, containing no duplicate characters
  """
  newStr = ''
  for ch in myString:
    if ch not in newStr:
      newStr = newStr + ch
  return newStr
```

##Listing 3.9 Remove the characters in one string from another

```
def removeMatches(myString, removeString):
  """
  Remove all characters matching a set of characters from a string.

  parameters
  ----------
  myString : string
    String to remove characters from
  removeString : string
    Set of characters to remove from myString

  returns 
  -------
    A new version of myString, with all characters in removeString removed
  """
  newStr = ''
  for ch in myString:
    if ch not in removeString:
      newStr = newStr + ch

  return newStr
```

##Listing 3.10 Generating a key starting from a password

```
def genKeyFromPass(password):
  """
  Make a key for the substitution cipher from a password.

  parameters
  ----------
  password : string
    password used to create the key

  returns
  -------
    substitution cipher key based on the password
  """
  from string import ascii_lowercase
  alphabet = ascii_lowercase + ' '
  password = password.lower()
  password = removeDupes(password)
  lastChar = password[-1]
  lastIdx = alphabet.find(lastChar)
  afterString = removeMatches(alphabet[lastIdx + 1:], password)
  beforeString = removeMatches(alphabet[:lastIdx], password)
  key = password + afterString + beforeString
  return key
```

##Listing 3.11 Looking up a letter in the Vigenère square

```
def vigenereIndex(keyLetter, plainTextLetter):
  """
  Use the Vigenere cipher to encipher a single character.

  parameters
  ----------
  keyLetter : string
    Single lowercase letter of the key
  plainTextLetter : string
    Single lowercase letter of the plaintext

  returns
  -------
    Enciphered version of the plaintext letter.
  """
  keyIndex = letterToIndex(keyLetter)
  ptIndex = letterToIndex(plainTextLetter)
  newIdx = (ptIndex + keyIndex) % 26
  return indexToLetter(newIdx)
```

##Listing 3.12 Encrypting a message using the Vigenère cipher

```
def encryptVigenere():
  """
  Function to get key and plaintext, then encipher with Vigenere cipher.
  """
  key = input('Enter the key: ')
  plainText = input('Enter the message to be encrypted: ')
  cipherText = ''
  keyLen = len(key)
  for i in range(len(plainText)):
    ch = plainText[i]
    if ch == ' ':
      cipherText = cipherText + ch
    else:
      cipherText = cipherText + vigenereIndex(key[i % keyLen], ch)
  return cipherText
```
