###1.1 Define the `checkWord()` Function

First, execute the cell below to define the `checkWord()` function, which is taken directly from Listing 8.11 of your text. This function takes a regular expression pattern as a parameter, and returns a list of all the dictionary words that match that pattern. 

Also, if needed, be sure to upload the `wordlist.txt` file, which is the list of valid dictionary words.

```
import re

def checkWord(regex):
  """
  Return a list of dictionary words matching a regular expression.

  parameters
  ----------
  regex : str
    Regular expression to match.

  returns
  -------
    A list of words matching the regular expression. 
  """
    
  resList = []
  with open('wordlist.txt', 'r') as wordFile:
    for line in wordFile:
      if re.match(regex, line[:-1]):
        resList.append(line[:-1])

  return resList
```
