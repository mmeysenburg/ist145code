##1 Accumulate!

The accumulator pattern is used to compute running sums or products. Soon we will be able to apply it to more realistic data, but for now, we will focus on computing sums or products of numbers that can be produced by the Python `range()` function.

###1.1 Accumulate and Average

Write a function named `avgN()` that uses the accumulator pattern to return the average of the first `n` positive integers, where `n` is a parameter. The first line of the function, and the docstring, are provided below. 

Think about doing this iteratively. First, come up with a `for` loop that iterates through the first `n` positive integers. Then, use the accumulator pattern to sum these numbers. Finally, determine the average, and send that value back with the `return` statement. 

````
def avgN(n):
  """
  Compute the average of the first n positive integers.

  parameters
  ----------
  n : int
    How many integers to include; assumed to be > 0

  returns
  -------
  The average of the first n positive integers, as a float.
  """
  # write the function definition here
  sum = 0
  for x in range(1, n + 1):
    sum = sum + x
  return sum / x
````
