##3 Celsius to Fahrenheit

It is easy to use Python to access weather forecasts and conditions from the [National Weather Service](https://www.weather.gov/) -- in fact, you will do that in Module 5! 

One issue is that the current temperature is reported in degrees Celsius, rather than degrees Fahrenheit. Suppose that the current temperature is 5.6 degrees Celsius, what is the temperature in Fahrenheit? Look up the conversion formula and write an expression in the cell below to determine the temperature in degrees Fahrenheit.

```
# Cell 1: degrees F for 5.6 degrees C
(9 / 5) * 5.6 + 32
```
