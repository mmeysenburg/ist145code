###2.4 Better Rectangles!

Now, write a *generalized* rectangle function, `drawRectangle()`, that takes three parameters: the turtle, the width, and the height. The function has been started for you in the cell below.

````
def drawRectangle(myTurtle, width, height):
  """
  This function draws a rectangle with the specified width and height.

  parameters
  ----------
  myTurtle : ColabTurtle.Turtle
    The turtle to draw with
  width : int
    The width of the rectangle
  height : int
    The height of the rectangle
  """
  myTurtle.forward(height)      # side 1
  myTurtle.right(90)
  myTurtle.forward(width)      # side 2
  myTurtle.right(90)
  myTurtle.forward(height)      # side 3
  myTurtle.right(90)
  myTurtle.forward(width)      # side 4
  myTurtle.right(90)  
````
