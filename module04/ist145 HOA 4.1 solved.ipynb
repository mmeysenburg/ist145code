###1.4 Write the `wordCount()` function

Now it's time to write your `wordCount()` function. Try to leverage what you've learned about `split()` into the function.

The first line of your function should be:

`def wordCount(sentence):`

````
# TODO: write your wordCount function here
def wordCount(sentence):
  """
  Count the number of words in a sentence.

  parameters
  ----------
  sentence : str
    String holding the sentence. Words are assumed to be separated by whitespace

  returns
  -------
    Number of whitespace-delimited words in the sentence.
  """
  sentenceList = sentence.split()
  return len(sentenceList)
````

###2.4 Write your `shuffleList()` function

Now you have the tools to write `shuffleList()`! Use the algorithm sketched in the comments in the code cell below.

````
# TODO: finish shuffleList() 
def shuffleList(myList):
  #make an empty list named returnList
  returnList = []

  #make a copy of myList named newList
  newList = myList[:]

  #remove elements from newList in random order, appending each to returnList
  for i in range(len(newList)):
    idx = random.randint(0, len(newList) - 1)
    val = newList.pop(idx)
    returnList.append(val)

  return returnList
````

###2.5 Test your `shuffleList()` function

Once you have written your `shuffleList()` function, test it by executing the code cell below. If your function fails any of the test cases, re-examine the requirements and your code to see if you can fix it. 

````
import collections
import random

testData = []
for i in range(10):
  newList = []
  for j in range(i):
    newList.append(random.randint(0, 100))
  newList.sort()
  testData.append(newList)

numCorrect = 0
for i, myList in enumerate(testData):
  print('Test case {0:d}:'.format(i + 1))
  print('\tOriginal:\t{0:s}'.format(str(myList)))
  newList = shuffleList(myList)
  print('\tShuffled:\t{0:s}'.format(str(newList)))
  if (len(myList) == len(newList)) and (collections.Counter(myList) == collections.Counter(newList)):
    numCorrect += 1
    print('\tPASSED!')
  else:
    print('\tFAILED')

if numCorrect == len(testData):
  print('Awesome!')
````
